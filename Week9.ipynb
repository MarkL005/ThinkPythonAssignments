```python
# ************************************************************
# Name: Mark Lucas
# Course: Intro to Python Programming
# Assignment: Midterm Project - Rental Car Pricing System
# Purpose: Calculate a customer's rental car bill based on
# car class, rental days, and miles driven.
# ************************************************************

# -------------------------
# Import Statements
# -------------------------

# No imports are needed for this program.

# -------------------------
# Function Definitions
# -------------------------

def calculate_bill(car_code, rental_days, start_odometer, end_odometer):
    miles_driven = end_odometer - start_odometer

    if car_code == "B":
        class_name = "Budget"
        daily_rate = 40.00
    elif car_code == "S":
        class_name = "Standard"
        daily_rate = 60.00
    else:
        class_name = "Luxury"
        daily_rate = 100.00

    if miles_driven > 100:
        charged_miles = miles_driven - 100
    else:
        charged_miles = 0

    mileage_charge = charged_miles * 0.35
    base_cost = rental_days * daily_rate
    total_bill = base_cost + mileage_charge

    return total_bill, miles_driven, charged_miles, class_name

# -------------------------
# Main Program
# -------------------------

print("Welcome to the Rental Car Pricing System")

car_code = input("Enter car classification code (B, S, or L): ").upper()

while car_code != "B" and car_code != "S" and car_code != "L":
    print("Invalid car code. Please enter B, S, or L.")
    car_code = input("Enter car classification code (B, S, or L): ").upper()

rental_days = int(input("Enter number of rental days: "))
start_odometer = float(input("Enter starting odometer reading: "))
end_odometer = float(input("Enter ending odometer reading: "))

total_bill, miles_driven, charged_miles, class_name = calculate_bill(
    car_code, rental_days, start_odometer, end_odometer
)

print()
print("Rental Summary")
print("Car class:", class_name)
print("Miles driven:", miles_driven)
print("Charged miles:", charged_miles)
print("Total bill: $" + format(total_bill, ".2f"))
```
