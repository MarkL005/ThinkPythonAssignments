```python
# Mark Lucas
# Intro to Programming with Python
# Week Assignment
# Team List Program

def add_player(team_list, player_name):
    team_list.append(player_name)

def search_player(team_list, player_name):
    if player_name in team_list:
        print(player_name + " is on the team.")
    else:
        print(player_name + " is not on the team.")

team_list = []

num_players = int(input("How many players do you want to add? "))

for i in range(num_players):
    player_name = input("Enter player name: ")
    add_player(team_list, player_name)

search_name = input("Enter a player name to search for: ")
search_player(team_list, search_name)

print("\nTeam List:")
for player in team_list:
    print(player)
```
