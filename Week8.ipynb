```python
##################################################
# Import Statements
##################################################


##################################################
# Function Definitions
##################################################

def is_palindrome(word):
    """
    Determine if the given word is a palindrome.
    Return True if it is.
    Return False if it is not.
    """

    # Convert the word to lowercase so comparison is case insensitive
    word = word.lower()

    # Reverse the word using string slicing
    reversed_word = word[::-1]

    # Compare the original lowercase word to the reversed word
    if word == reversed_word:
        return True
    else:
        return False


##################################################
# Program Logic
##################################################

# Ask the user to enter a word
user_word = input("Enter a word: ")

# Call the is_palindrome function and store the result
result = is_palindrome(user_word)

# If the result is True
if result:
    print(f"{user_word} IS a palindrome.")
# Otherwise
else:
    print(f"{user_word} is NOT a palindrome.")
```
